```python
# data to plot
cat_id = list(df_time_delta.category_id)
trend_time = list(df_time_delta.time_delta)

objects = []
for cat in cat_id:
    for index, name in cat_data.items():
        if index == int(cat):
            objects.append(name)

plt.figure(figsize=(12,6))
sns.barplot(y=trend_time, x=objects, alpha=0.6)
plt.xticks(rotation=90,fontsize=20)
plt.yticks(fontsize=20)
plt.title("How Long Does It Take for a Video to Become Trending?", fontsize=20,y=1.06)
plt.ylabel("# of Days", fontsize=20)
plt.savefig('photos/avg_days_to_trend.png', bbox_inches='tight', dpi=100)
plt.show()
```

## Avg # of days a video is trending

```python
df_tot_trend_days = df.groupby(['category_id','video_id']).count()
df_tot_trend_days = df_tot_trend_days[['trending_date']]
df_tot_trend_days
```

check whether the video with id kgaO45SyaO4 is counted three times as it trended for 3 days

```python
df_tot_trend_days.loc[(28.0,'kgaO45SyaO4')]
```

```python
df_tot_trend_days = df_tot_trend_days.reset_index(level='video_id')
df_tot_trend_days
```

```python
df_avg_trend_days = df_tot_trend_days.groupby('category_id')['trending_date'].mean()
df_avg_trend_days = df_avg_trend_days.sort_values(ascending=False)
df_avg_trend_days
```

```python
# data to plot
cat_id = list(df_avg_trend_days.index)
avg_trend_day = list(df_avg_trend_days)

objects = []
for cat in cat_id:
    for index, name in cat_data.items():
        if index == int(cat):
            objects.append(name)

plt.figure(figsize=(12,6))
sns.barplot(y=avg_trend_day, x=objects, alpha=0.6)
plt.xticks(rotation=90,fontsize=20)
plt.yticks(fontsize=20)
plt.title("How Long Does a Video Stay Trending?", fontsize=20, y=1.06)
plt.ylabel("# of Days", fontsize=20)
plt.savefig('photos/avg_trending_days.png', bbox_inches='tight', dpi=100)
plt.show()
```

## Comparison: Avg # of days to become trending VS. avg # of trending days

```python
df_combined = pd.merge(df_time_delta[['category_id','time_delta']], pd.DataFrame({'category_id':df_avg_trend_days.index, 'trending_days':df_avg_trend_days.values}), on='category_id', how='right')
df_combined = df_combined.sort_values(by='time_delta', ascending=False)
df_combined
```

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 10))

cat_id = list(df_combined.category_id)
time_del = list(df_combined.time_delta)
trending_days = list(df_combined.trending_days)

objects = []
for cat in cat_id:
    for index, name in cat_data.items():
        if index == int(cat):
            objects.append(name)

r1 = np.arange(len(objects))
r2 = [x + 0.3 for x in r1]

# Create blue bars
plt.bar(r1, time_del, width = 0.3, color = 'blue', edgecolor = 'black', capsize=7,
        label='Average # of Days It Takes For a Video to Become Trending')
 
# Create cyan bars
plt.bar(r2, trending_days, width = 0.3, color = 'cyan', edgecolor = 'black', capsize=7,
        label='Average # of Days A Video Is Trending')
 
# general layout
plt.xticks([r + 0.15 for r in range(len(objects))], objects, rotation = 70)
plt.ylabel('# of Days')
plt.title("# of Days to Become Trending VS. # of Days of Trending", fontsize=16)
plt.legend()
plt.savefig('photos/comp_trending.png', bbox_inches='tight', dpi=100)
# Show graphic
plt.show()

```

```python
df_dup_diff = df_date.sort_values('video_id', kind='mergesort')
del df_dup_diff['Unnamed: 0'],df_dup_diff['description'],df_dup_diff['tags']
df_dup_diff = df_dup_diff.reset_index()
df_dup_diff
```

```python
df_dup_diff[['view_diff','like_diff','dislike_diff','comment_diff']] = pd.DataFrame(np.nan, index=df_dup_diff.index, columns=['view_diff','like_diff','dislike_diff','comment_diff'])
df_dup_diff
```

```python
for index, row in df_dup_diff.iterrows():
    if index==0:
        df_dup_diff.at[index,'view_diff'] = df_dup_diff.at[index,'views']
        df_dup_diff.at[index,'like_diff'] = df_dup_diff.at[index,'likes']
        df_dup_diff.at[index,'dislike_diff'] = df_dup_diff.at[index,'dislikes']
        df_dup_diff.at[index,'comment_diff'] = df_dup_diff.at[index,'comment_count']
    else:
        if df_dup_diff.at[index,'video_id'] == df_dup_diff.at[index-1,'video_id']:
            df_dup_diff.at[index,'view_diff'] = df_dup_diff.at[index,'views'] - df_dup_diff.at[index-1,'views']
            df_dup_diff.at[index,'like_diff'] = df_dup_diff.at[index,'likes'] - df_dup_diff.at[index-1,'likes']
            df_dup_diff.at[index,'dislike_diff'] = df_dup_diff.at[index,'dislikes'] - df_dup_diff.at[index-1,'dislikes']
            df_dup_diff.at[index,'comment_diff'] = df_dup_diff.at[index,'comment_count'] - df_dup_diff.at[index-1,'comment_count']
        else:
            df_dup_diff.at[index,'view_diff'] = df_dup_diff.at[index,'views']
            df_dup_diff.at[index,'like_diff'] = df_dup_diff.at[index,'likes']
            df_dup_diff.at[index,'dislike_diff'] = df_dup_diff.at[index,'dislikes']
            df_dup_diff.at[index,'comment_diff'] = df_dup_diff.at[index,'comment_count']

df_dup_diff
```
