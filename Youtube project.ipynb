```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
csvpath='/Users/lauradu/Downloads/youtube-new/USvideos.csv'
Youtube_df=pd.read_csv (csvpath)
Youtube_df.head()
```

```python
Youtube_df.count()
```

```python
Youtube_df.dtypes
```

```python
clean_Youtube_df = Youtube_df.dropna(how="any")
clean_Youtube_df.count()
```

```python
category_Youtube_df=clean_Youtube_df.loc[clean_Youtube_df["category_id"] == 22, :]
category_Youtube_df
```

```python
Category_Youtube_df = Youtube_df.groupby('category_id').sum()
Category_Youtube_df.head()
```

```python
Category_Youtube_df=Category_Youtube_df.reset_index()
Category_Youtube_df.head()
```

```python
Category_Youtube_df['Percentage_likes'] =Category_Youtube_df.likes/Category_Youtube_df.views*100
Category_Youtube_df.head()
```

```python
Category_Youtube_df['Percentage_dislikes'] =Category_Youtube_df.dislikes/Category_Youtube_df.views*100
Category_Youtube_df.head()
```

```python
Category_Youtube_df.shape
```

```python
Category_Youtube_df = Category_Youtube_df.sort_values(by='Percentage_likes', ascending=False)
Category_Youtube_df
```

```python
import seaborn as sns

cat_data = {2:'Autos & Vehicles',
           1:'Film & Animation',
           10:'Music',
           15:'Pets & Animals',
           17:'Sports',
           18:'Short Movies',
           19:'Travel & Events',
           20:'Gaming',
           21:'Videoblogging',
           22:'People & Blogs',
           23:'Comedy',
           24:'Entertainment',
           25:'News & Politics',
           26:'Howto & Style',
           27:'Education',
           28:'Science & Technology',
           29:'Nonprofits & Activism',
           30:'Movies',
           31:'Anime/Animation',
           32:'Action/Adventure',
           33:'Classics',
           34:'Comedy',
           35:'Documentary',
           36:'Drama',
           37:'Family',
           38:'Foreign',
           39:'Horror',
           40:'Sci-Fi/Fantasy',
           41:'Thriller',
           42:'Shorts',
           43:'Shows',
           44:'Trailers'}
cat_id = list(Category_Youtube_df['category_id'])
objects = []
for cat in cat_id:
   for index, name in cat_data.items():
       if index == int(cat):
           objects.append(name)

plt.figure(figsize=(10,6))
sns.barplot(y=Category_Youtube_df['Percentage_likes'], x=objects, alpha=0.6)
plt.xticks(rotation=60)
plt.title("Youtube Percentage of Likes", fontsize=16)
plt.xlabel("ID", fontsize=16)
plt.ylabel("Video Percentage Likes", fontsize=16)
plt.show()
```

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(13, 10))

# Plot the total schools per city
sns.set_color_codes("pastel")
sns.barplot(x=objects, y='Percentage_likes', data=Category_Youtube_df,
           label="likes percentage", color="b")
plt.xticks(rotation=60)

# Plot the total community schools per city
sns.set_color_codes("muted")
sns.barplot(x=objects, y="Percentage_dislikes", data=Category_Youtube_df,
           label="dislikes percentage", color="r")
plt.legend()
```
